## Analiza głównych składowych (PCA)

PCA obraca układ współrzędnych tak, aby maksymalizować w pierwszej kolejności wariancję pierwszej współrzędnej, następnie wariancję drugiej współrzędnej itd.. W ten sposób konstruowana jest nowa przestrzeń obserwacji, w której najwięcej zmienności wyjaśniają początkowe czynniki. 

<div>

</div>

Metoda polega na znalezieniu wartości własnych i wektorów własnych macierzy kowariancji danych. Wektor własny o największej wartości własnej określa kierunek (składową) PCA o największej wariancji. Gdy wybierzemy $k$ składowych o największych wartościach własnych to mogą one posłużyć do rzutowania danych $n$-wymiarowych do przestrzeni $k$-wymiarowej. 

### Algorytm PCA:
1. wycentrowanie danych (wyśrodkowanie wartości dla każdej zmiennej)  
   $$ x_{ij} \leftarrow x_{ij} - \frac{1}{n} \sum_{j=1}^n x_{ij}$$ 
2. wyznaczenie macierzy kowariancji
  $$ \mathbf{R} = \frac{1}{n} \mathbf{X}\mathbf{X}^T  \quad \text{gdzie} \quad \mathbf{X} = \{ x_{ij}\}$$
3. rozkład macierzy kowariancji na wektory i wartości własne, np.  korzystając z rozkładu SVD 
$$ \mathbf{R} \mathbf{w}_i = \lambda_i \mathbf{w}_i \qquad i=1,\ldots, N $$
$$\mathbf{S} = \mathbf{W}^{-1}\mathbf{R}\mathbf{W} $$
gdzie $\mathbf{W}$ zawiera wektory własne a $\mathbf{S}$ wartości własne na diagonali   
4. wybór $k$ składowych o największych wartościach własnych
5. utworzenie macierzy rzutowania $\mathbf{W}_k$ składającej się z wybranych $k$ wektorów własnych
6. przekształcenie danych do $k$ wymiarowej przestrzeni $ \mathbf{Y}= \mathbf{W}_k \mathbf{X}  $

````python
import numpy as np

X_c = X - np.mean(X, axis=0)           # centrowanie danych

n = X_c.shape[0]                       # liczba przypadkow      
R = X_c.T @ X_c / n                    # macierz kowariancji
# to samo R = np.cov(X.T)

# wektory własne i wartości własne macierzy kowariancji
eigen_vals, eigen_vecs = np.linalg.eig(R)

print('Wartości własne\n', eigen_vals)
print('Wektory własne\n', eigen_vecs)
````

### Pierwsza składowa główna PC1

Składowa o największej wartości własnej

````python
print('Największa wartość własna ', eigen_vals[0])
pc1 = eigen_vecs.T[0]
x1 = X_c @ pc1
print('Wariancja wzdłóż kierunku PC1 ', np.var(x1))
print('Procent wyjaśnianej wariancji ', 100.0 * eigen_vals[0]/ np.sum(eigen_vals))
````

````python
plt.plot(X_c[:,0],X_c[:,1], 'o')
plt.arrow(0, 0, pc1[0], pc1[1], width=0.05, color='red', label='PC1')
plt.legend();
````

## Reguła Oji

Perceptron liniowy 
$$ y = \mathbf{w}\mathbf{x}$$
Reguła uczenia Hebba
$$\Delta \mathbf{w} = \eta (y \mathbf{x} - \alpha y^2 \mathbf{w})$$

gdzie $\alpha > 0$ wpływa na wartość normy  $ ||\mathbf{w}|| = \frac{1}{\alpha}$ 

### Ćwiczenie:

````python
## rozwiązanie

class OjaPerceptron(HebbPerceptron):
    
    def __init__(self, n_features=1, eta=0.01, n_epochs=100, alpha=1):
        self.eta = eta
        self.n_features = n_features
        self.n_epochs = n_epochs
        self.alpha = alpha
    
    def fit(self, X):
        self.init(X)
        self.weights.append(self.W[0].copy())
        for _ in range(self.n_epochs):
            for x in np.random.permutation(X):
                y = self.transform(x)
                dw = self.eta*( (y*x) - (self.alpha*y*y*self.W) )
                self.W += dw
                self.weights.append(self.W[0].copy())

       # TODO
        return self
````

## Zadanie: Ogólny algorytm Hebbowski (GHA)


Zaimplementuj sieć jednowarstwową realizującą rozkład na czynniki główne za pomocą ogólnego algorytmu hebbowsiego (GHA). W implementacji możesz wykorzytsać szablon klasy ``GHA`` umieszczony w pliku [gha.py](gha.py).

**Parametry początkowe** (ustawiane w argumentach konstruktora):
* $N$ liczba epok
* $k$ liczba neuronów (komponentów)
* $\eta$ wspólczynnik uczenia

**Algorytm uczenia GHA** (realizowany w metodzie `fit(X)`)  

1. Zainicjuj macierz wag $\mathbf{W}$ 
2. Powtarzaj $N$ razy:
3. <ul>Dla każdego przypadku $\mathbf{x}$ ze zbioru uczącego wykonaj</ul>
4. <ul><ul> wyznacz aktywacje <br> $\mathbf{y} = \mathbf{W}\mathbf{x}$ </ul></ul>
5. <ul><ul>zaktualizuj wagi <br> $\Delta \mathbf{W}=\eta \left(\mathbf{y} \cdot \mathbf{x}^T-\mathrm{LT}\left[\mathbf{y} \cdot \mathbf{y}^T\right] \mathbf{W}\right)$ </ul></ul>
gdzie operacja $\mathbf{a}\cdot \mathbf{b}^T$ oznacza iloczyn zewnętrzny, a funkcja $\mathrm{LT}[\mathbf{A}]$ zeruje wszystkie elementy macierzy $\mathbf{A}$ znajdujące się powyżej diagonali

Przydatne funkcje:
* iloczyn zewnętrzny można wyznaczyć za pomocą funkcji [numpy.outer(a,b)](https://numpy.org/doc/stable/reference/generated/numpy.outer.html),
* funkcja [numpy.tril(x)](https://numpy.org/doc/stable/reference/generated/numpy.tril.html) zeruje wartości powyżej diagonali macierzy.

````python
from sklearn.datasets import fetch_openml
import numpy as np

mnist = fetch_openml('mnist_784', data_home='./dane/', parser='auto')

n = 1000   # wybieramy 1000 przypadków
ind = np.random.permutation(mnist.data.shape[0])

X = mnist.data.values[ind[:n]]
target =  mnist.target.values[ind[:n]]
w, h = 28, 28   # rozmiary obrazów

print('X.shape :', X.shape)
print('target: ', np.unique(target))

# normalizacja danych
X_norm = 2.0* X / 255.0 - 1.0

print(f'Min {X_norm.min()}, max {X_norm.max()}')

````

````python
import numpy as np

class GHA(object):
    
    def __init__(self, n_components=1, eta=0.0001, n_epochs=100):
        self.eta = eta
        self.n_components = n_components
        self.n_epochs = n_epochs
        
    def init(self, X):
        # TODO inicjuj macierz wag self.W o kształcie [n_components, ilość zmiennych ] 
        self.weights = []
        self.W = np.random.rand(self.n_components, X.shape[1])
        return self

    def fit(self, X):
        # TODO algorytm uczenia GHA
        self.init(X)
        self.weights.append(self.W[0].copy())
        for _ in range(self.n_epochs):
            for x in np.random.permutation(X):
                y = self.transform(x)
                dw = self.eta * (np.outer(y,x.T) - np.tril(np.outer(y,y.T)) @ self.W)
                self.W += dw
                self.weights.append(self.W[0].copy())
        return self
            
    def transform(self, X):
        return (self.W @ X.T).T

    def inverse_transform(self, Y):
        return Y @ self.W
````

````python

n_components = 100
eta = 0.0000001
epchos=100
pca = GHA(n_components=n_components,eta=eta,n_epochs=epchos)
pca.fit(X_norm)
X_norm_pca = pca.transform(X_norm)
print(X_norm_pca)
print('Kształt zbioru danych przed transformacją:', X_norm.shape)
print('Kształt zbioru danych po transformacji:', X_norm_pca.shape)
````
